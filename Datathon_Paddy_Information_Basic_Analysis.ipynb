Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

Import data and Data Cleaning/Preprocessing

```python
df = pd.read_csv("paddydataset.csv")

print("\nFirst 5 rows")
print(df.head())
print("\nLast 5 rows")
print(df.tail())

print("\nShape of dataset")
print(df.shape)

print("\nColumns of dataset")
print(df.columns)

print("\nDatatype of dataset")
print(df.dtypes)
```

```python
print("\nCheck for missing values: ")
print(df.isnull().sum())

print("\nCheck for duplicates: ")
print(df.duplicated().sum())
```

```python
print("Remove duplicate:")
print(df.drop_duplicates(inplace=True))

print("\nCheck for duplicates: ")
print(df.duplicated().sum())
```

```python
print(df.head())
```

Exploratory Data Analysis (EDA)

```python
print("\nDescribe of dataset")
print(df.describe())

print("\nInfo of dataset")
print(df.info())
```

```python
print("Label Encoding")
le = LabelEncoder()
df["Agriblock_label"] = le.fit_transform(df["Agriblock"])
df
```

Visualisation (Graphs/Charts)

```python
sns.pairplot(df)
df
```

```python
plt.figure(figsize=(10, 10))
print("Scatter plot")
sns.scatterplot(x="Agriblock",y="Paddy yield(in Kg)",data=df)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
print("Line plot")
x=df["Agriblock"]
y=df["Paddy yield(in Kg)"]
plt.plot(x,y)
plt.xlabel("Agriblock")
plt.ylabel("Paddy yield(in Kg)")
plt.title("Line plot")
plt.show()
```

```python
sns.lineplot(x="Agriblock",y="Paddy yield(in Kg)",data=df)
```

```python
plt.figure(figsize=(10, 10))
print("Scatter plot")
sns.scatterplot(x="Paddy yield(in Kg)",y="Soil Types",data=df)
plt.show()
```

```python
plt.bar(df["Agriblock"],df["Paddy yield(in Kg)"])
```

```python
plt.bar(df["Soil Types"],df["Paddy yield(in Kg)"])
```
